```python
# this code calculates how many days it takes to reach a target number by incrementally adding numbers starting from 1, with a multiplier affecting the frequency of increments.

target = int(input("Enter target number: "))
multiplier = int(input("1 is hardest to the number you can put higher is the easiest: "))

total = 0
count = 0
current = 1
times_used = 0

for _ in range(target + 1):  # Upper bound - worst case every addition is 1
    if total >= target:
        break
    
    total += current
    count += 1
    times_used += 1
    
    if times_used == (multiplier * current):
        current += 1
        times_used = 0

print(f"Total: {total}, Last number: {current - 1 if times_used == 0 else current}, Days: {count}, Iteration: {times_used}")
```

```python
# this code calculates how many days it takes to reach a target number by cycling through numbers 1 to max_number

target = int(input("Enter target number: "))
max_number = int(input("Enter the highest number to use (e.g., 3 means cycle through 1, 2, 3): "))

total = 0
count = 0
current = 1
pattern = []

while total < target:
    total += current
    count += 1
    pattern.append(str(current))
    
    # Move to next number, cycling back to 1 after max_number
    current += 1
    if current > max_number:
        current = 1

# Calculate the last number that was actually used
last_used = current - 1 if current > 1 else max_number
pattern.sort(key=int)  # Sort pattern for better readability
print(f"Pattern: {''.join(pattern)}")

# Count occurrences of each number
for i in range(1, max_number + 1):
    num_count = pattern.count(str(i))
    print(f"{i}s: {num_count}")

print(f"Total: {total}, Last number used: {last_used}, Days: {count}")
```

```python
# this code calculates days to reach target where each number N appears N times (1 once, 2 twice, 3 thrice, etc.)

target = int(input("Enter target number: "))
max_number = int(input("Enter the highest number to use (e.g., 3 means 1, 22, 333): "))

total = 0
count = 0
current = 1
times_used = 0
pattern = []

while total < target:
    total += current
    count += 1
    pattern.append(str(current))
    times_used += 1
    
    # Move to next number after using current number 'current' times
    if times_used == current:
        current += 1
        times_used = 0
        # Loop back to 1 after max_number
        if current > max_number:
            current = 1

# Calculate the last number that was actually used
last_used = int(pattern[-1]) if pattern else 0
pattern.sort(key=int)  # Sort pattern for better readability
print(f"Pattern: {''.join(pattern)}")

# Count occurrences of each number
for i in range(1, max_number + 1):
    num_count = pattern.count(str(i))
    print(f"{i}s: {num_count}")

print(f"Total: {total}, Last number used: {last_used}, Days: {count}")
```
